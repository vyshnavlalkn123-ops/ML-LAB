```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'Area': [1200, 1500, 1800, 2400, 3000, 3500, 4000, 4500, 5000, 5500],
    'Bedrooms': [2, 3, 3, 4, 5, 5, 6, 6, 7, 7],
    'Age': [10, 5, 8, 2, 1, 3, 2, 4, 1, 2],
    'Distance': [5, 3, 4, 2, 1, 2, 1, 3, 2, 1],
    'Price': [250, 320, 340, 450, 600, 650, 700, 720, 800, 850]
}
df = pd.DataFrame(data)

print(df)
```

```python
X = df[['Area', 'Bedrooms', 'Age', 'Distance']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
print("Intercept:", model.intercept_)

print("\nCoefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(feature, ":", coef)
```

```python
print("\nRegression Equation:")
print("Price =", end=" ")
for feature, coef in zip(X.columns, model.coef_):
    print(f"({coef:.4f} * {feature}) + ", end="")
print(model.intercept_)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("\nMean Squared Error:", round(mse, 4))
print("R2 Score:", round(r2, 4))
```

```python
new_house = pd.DataFrame({
    'Area': [2800],
    'Bedrooms': [4],
    'Age': [2],
    'Distance': [2]
})
```

```python
predicted_price = model.predict(new_house)
```

```python
print("\nNEW HOUSE PREDICTION")

print("\nPredicted House Price:", round(predicted_price[0], 2))

plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.grid(True)
plt.show()
```
